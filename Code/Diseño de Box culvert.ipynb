# Diseño de Box Culvert

```python
import numpy as np 
from math import asin, sqrt, pow, ceil , floor
import matplotlib.pyplot as plt
import sympy as sym
from sympy import integrate, init_printing
from sympy.abc import x
init_printing(use_latex="mathjax")
from sympy.abc import a,b,c
from scipy.interpolate import interp1d
import math

m = 1                 
kN = 1          
sec = 1         
mm = m/1000.             
cm = m/100.
N = kN/1000.
MPa = N/mm**2
GPa = MPa*1000
m2 = m*m           
m3 = m*m*m           
m4 = m*m*m*m
inch = cm*2.54

# Datos de entrada

H_r= 1.5*m   # Altura del relleno
Gsoil= 18*kN/m3  # Gamma del suelo
Gadm=150*kN/m2   # Gamma admisible del suelo
fi=28 # Ángulo de fricción
B=4*m   # Base de una celda del box culvert
H=3*m    # Altura del box culvert
Kb=34500*kN/m3   # Coeficiente de balastro
```

```python
K0=1-math.sin(fi)  # Coeficiente pasivo
print('K0 = %5.2f' %(K0))
Ka=math.tan(45-fi/2)**2   # Coeficiente activo
print('Ka = %5.2f' %(Ka))
```

```python
p_sup=K0*Gsoil*H_r  # Presión ejercida por el suelo en la parte superior del box culvert 
p_inf=K0*Gsoil*(H_r+H) # Presión ejercida por el suelo en la parte inferior del box culvert 
print('p_sup = %5.2f kN/m' %(p_sup))
print('p_inf = %5.2f kN/m' %(p_inf))
```
